```python
def get_filenames(folder_directory): #retrieves all files in directory and subdirectories starting with specified characters (keyword_start)
    allfiles=list()
    titles=list()
    for file in os.listdir(folder_directory):
        if file.endswith(".eis"):
            titles.append(file.split('.',1)[0])
            allfiles.append(os.path.join(folder_directory, file))
    return allfiles, titles
```
