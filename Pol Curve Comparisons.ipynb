```python
#Script Purpose: Plots Pol curve from .pol folder to compare
#Author: Shan Kutagulla
import os
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import openpyxl
import numpy as np
import scipy
```

```python
def get_filenames(folder_directory): #retrieves all files in directory and subdirectories starting with specified characters (keyword_start)
    allfiles=list()
    titles=list()
    for file in os.listdir(folder_directory):
        if file.endswith(".pol"):
            titles.append(file.split('.',1)[0])
            allfiles.append(os.path.join(folder_directory, file))
    return allfiles, titles
```

```python
pol_list,titles=get_filenames("/Users/shanmukhkutagulla/My Drive/Fuel Cell Docs/Pol Curve Processed Files/Working Pol Data")
```

```python
plt.figure(figsize=(8, 6), dpi=100)
for file in pol_list:
    df=pd.read_csv(file)
    plt.errorbar(df["Average Pol I (A/cm2)"],df["Average Pol E_Stack (V)"],xerr=[df["Average Downwards X Error"],df["Average Upwards X Error"]])
plt.ylabel("E_Stack (V)")
plt.xlabel("Average Pol I (A/cm2)")
plt.legend(titles)
```
